```python
import pandas as pd
data = pd.read_csv('proyeksi-penduduk-2010-2035.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data
```

Data dari 34 provinsi di Indonesia dari tahun 2010 dan diramalkan jumlah penduduk hingga tahun 2035. Proyeksi hingga tahun 2035 dibagi menjadi 2 jenis kelamin laki-laki dan perempuan.

```python
DKI = data[data.propinsi == "DKI Jakarta"]
```

```python
DKI
```

```python
JB = data[data.propinsi == "Jawa Barat"]
```

```python
JB
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot (JB.tahun, JB.jumlah)
plt.plot (DKI.tahun, DKI.jumlah)
plt.legend(["Jawa Barat","DKI Jakarta"])
plt.xlabel("Tahun")
plt.ylabel("jumlah penduduk")
plt.show
```

```python
import statistics as stat
a = stat.mean(DKI.jumlah)
b = stat.mean (JB.jumlah)
print("rata-rata penduduk DKI Jakarta adalah", a, "sedangkan rata-rata pendukduk Jawab Barat adalah", b, "Sehingga penduduk jawa barat lebih banyak dari jakarta")
```

```python
c = stat.median(DKI.jumlah)
d = stat.stdev(DKI.jumlah)
e = stat.variance(DKI.jumlah)
f = stat.median(JB.jumlah)
g = stat.stdev(JB.jumlah)
h = stat.variance(JB.jumlah)

print("DKI Jakarta :\n===========")
print("median :", c)
print("stdev :",d)
print("variance :",e,"\n==========")
print("Jawa Barat :\n===========")
print("median :", f)
print("stdev :",g)
print("variance :",h,"\n==========")
```

##Insight yang diperoleh adalah data Jawa Barat dari segi median, stdev, dan varian lebih besar dari DKI Jakarta
